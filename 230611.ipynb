```python
# 0을 포함하지 않는 야구게임 ver2
import random
while True:
    print("0을 포함하지 않는 야구게임\n")
    print("**게임규칙**")
    print("이 게임은 컴퓨터에서 임의로 지정한 3자리 숫자를 맞추는 게임입니다.")
    print("각 회별로 3자리의 숫자를 입력하실 수 있으며")
    print("3자리의 숫자 중 숫자는 맞지만 위치가 틀렸을 때는 볼로")
    print("숫자와 위치가 전부 맞으면 스트라이크로 알려줍니다")
    print("그리고 3자리의 숫자가 숫자, 위치를 전부 틀리면 3아웃으로 표시됩니다.")
    print("이 게임에서는 중복과 숫자 0이 포함되지 않으며")
    print("어느자리가 스트라이크 또는 볼인지 알려주지는 않습니다.\n")
    start = input("게임을 시작하시려면 키보드의 ENTER를 눌러주세요.")
    while True:
        # 리스트, 변수 초기화
        baseball_list = []
        user_list = []
        strike = 0
        ball = 0
        # 시도 횟수
        num = 0
        # random함수로 123~987 사이의 숫자를 뽑기
        # random.randrange(m,n+1) : m부터 n까지의 숫자를 랜덤으로 뽑기
        baseball = random.randrange(123,988)
        # 1. 백의자리는 0이 나올 수 없기에 십의자리랑 일의자리에 0이 나오지 않을 때 계속 진행
        if ((baseball%100)//10!=0) and ((baseball%10)!=0):
            # 2. 백의자리와 십의자리 숫자, 백의자리와 일의자리 숫자, 십의자리와 일의자리 숫자가 같아서는 안됨
            if (((baseball//100)!=((baseball%100)//10)) and ((baseball//100)!=(baseball%10)) and (((baseball%100)//10)!=(baseball%10))):
                # baseball_list에 랜덤으로 뽑은 숫자의 백,십,일의자리 숫자 넣음
                for i in range(3):
                    baseball_list.append(int(str(baseball)[i]))
                while True:
                    # 시도 추가
                    num = num +1
                    if (num==9):
                        print("게임에서 지셨습니다.")
                        break
                    else:
                        # print("....")
                        pass
                    # 3. 게임 진행할 숫자(0을 제외한) 뽑고 중복 제거
                    print("="*47)
                    user = input(f"{num}회\n사용자 입력숫자 : ")
                    if (len(user)==3):
                        if (((int(user[0])!=0) and (int(user[1])!=0) and (int(user[2])!=0)) and ((int(user[0]) != int(user[1])) and (int(user[0]) != int(user[2])) and (int(user[1]) != int(user[2])))):
                            for i in range(3):
                                user_list.append(int(user[i]))
                        else:
                            continue
                    else:
                        continue
                    # 4. 스트라이크 계산
                    for i in range(3):
                        if (baseball_list[i]==user_list[i]):
                            strike = strike+1
                        #  5. 볼 계산
                        elif baseball_list[i] in user_list:
                            ball = ball+1
                    # 3스트라이크면 break로 게임 승리
                    if (strike==3):
                        print(f"결과 : {strike}스트라이크 게임 승리")
                        break
                    # 3아웃이면 contuinue로 while문 다시 시작
                    elif ((strike==0) and (ball==0)):
                        print(f"결과 : 3 OUT!!!")
                        continue
                    # 나머지는 continue로 while문 다시 시작
                    else:
                        print(f"결과 : {strike}스트라이크 {ball}볼")
                        user_list = []
                        strike=0
                        ball=0
                        continue
            else:
                # print("숫자를 다시 뽑아야함")
                continue
            break
    retry = input("다시 하시겠습니까? (Y/N)")
    if (retry == 'Y'):
        print("다시 ㄱㄱ")
        continue
    elif (retry == 'N'):
        print("게임 종료")
        break
    print("잘못 입력하셔서 게임이 종료됩니다.")
    break
```
