````python
## ※  문자열 포매팅

- “현재 온도는 18도입니다.”, “현재 온도는 20도입니다.” ⇒ 숫자의 차이
    
    → 문자열 안의 특정한 값을 바꿔야 할 때 문자열 포매팅 필요
    
1. 숫자 바로 대입

```python
# %d : 정수 대입
>>> "I eat %d apples." %3
'I eat 3 apples'
```

1. 문자열 바로 대입

```python
# %s : 문자열 대입
>>> "I eat %s apples." % "five"
'I eat five apples.'
```

1. 숫자 값을 나타내는 변수로 대입

```python
# 1번처럼 숫자를 대입하는데 숫자를 변수에 넣어서 대입
>>> number = 3
>>> "I eat %d apples." % number
'I eat 3 apples'
```

1. 2개 이상의 값 넣기

```python
# 2개 이상의 값을 포맷팅으로 대입하기 위해서는
# 뒤의 %에 넣을 값들을 괄호안에 콤마로 값 넣어주면 됨
>>> number = 10
>>> day = "three"
>>> 'I ate %d apples. so I was sick for %s days.' % (number, day)
'I ate 10 apples. so I was sick for three days.'
```

1. 문자열 포맷 코드

```python
# %s : 문자열
# %c : 문자 1개
# %d : 정수
# %f : 부동 소수(실수)
# %o : 8진수
# %x : 16진수
# %% : % 기호 자체를 쓰기 위함
```

1. 포맷 정렬과 공백

```python
>>> "%10s" % "hi"
'        hi'
```

→ 전체 길이가 10칸 중 문자열을 오른쪽 정렬로 대입(즉, 위의 예제에서는 10칸 중 8칸을 공백으로 하고 hi가 됨)

```python
>>> "%-10sjane" % 'hi'
'hi        jane'
```

→ jane 앞으로 10칸 중 문자열(hi)이 왼쪽 정렬로 대입

1. 소수점 표현

```python
>>> "%0.4f" % 3.42134234
'3.4213'
```

→ %0.4f : 소수점 4자리까지 표현

```python
>>> "%10.4f" % 3.42134234
'    3.4213'
```

→ 소수점 4자리까지만 표시한 수를 10칸인 공간에서 오른쪽 정렬

1. format 함수를 사용한 포매팅

```python
# 숫자 바로 대입({0}이 숫자 3으로 바뀜)
>>> 'I eat {0} apples".format(3)
'I eat 3 apples.'

# 문자열 바로 대입({0}이 문자열 five로 바뀜)
>>> "I eat {0} apples".format("five")
'I eat five apples.'

# 숫자 값을 가진 변수로 대입하기({0}이 변수 number의 값인 3으로 바뀜)
>>> number = 3
>>> "I eat {0} apples".format(number)
'I eat 3 apples.'

# 2개 이상의 값 넣기({0},{1}이 number,day 변수의 값인 10과 three로 변함)
>>> number = 10
>>> day = "three"
>>> "I ate {0} apples. so I was sick for {1} days".format(number, day)
'I ate 10 apples. so I was sick for three days'

# 인덱스와 이름을 혼용해서 넣기(인덱스 항목과 name=value 형태 혼용 가능)
>>> "I ate {0} apples. so I was sick for {day} days.".format(10, day=3)
'I ate 10 apples. so I was sick for 3 days'

# 왼쪽 정렬(총 자리 10칸 중 포맷팅하는 문자를 왼쪽 정렬)
>>> "{0:<10}".format("hi")
'hi        '

# 오른쪽 정렬(총 자리 10칸 중 포맷팅하는 문자를 오른쪽 정렬)
>>> "{0:>10}".format("hi")
'        hi'

# 가운데 정렬(총 10칸에서 문자를 가운데 정렬)
>>> "{0:^10}".format("hi")
'    hi    '

# 공백 채우기(공백을 문자로 채우기)
>>> "{0:=^10}".format("hi")
'====hi===='
>>> "{0:!<10}".format("hi")
'hi!!!!!!!!'

# 소수점 표현하기(아래 예제에서는 소수점 4자리까지 표현)
>>> y = 3.42134234
>>> "{0:0.4f}".format(y)
'3.4213'

# { 또는 } 문자 표현하기(*)
# 중괄호 안의 문자를 문자 그대로 사용하고 싶을 경우
>>> "{{ and }}".format()
'{ and }'

## f 문자열 포매팅(문자열 앞에 f를 붙이면 사용 가능)
# 파이썬 3.6 미만 버전에서 사용 불가
>>> name = '홍길동'
>>> age = 30
>>> f'나의 이름은 {name}입니다. 나이는 {age}입니다.'
'나의 이름은 홍길동입니다. 나이는 30입니다.'

>>> age = 30
>>> f'나는 내년이면 {age+1}살이 된다.'
'나는 내년이면 31살이 된다.'
# 딕셔너리도 사용 가능
>>> d = {'name':'홍길동', 'age':30}
>>> f'나의 이름은 {d["name"]}입니다. 나이는 {d["age"]}입니다.'
'나의 이름은 홍길동입니다. 나이는 30입니다.'
# 정렬, 공백에서도 사용 가능
# 왼쪽 정렬
>>> f'{"hi":<10}'
'hi        '
# 가운데 정렬에 공백을 =으로 채우기
>>> f'{"hi":=^10}'
'====hi===='
# 소수점
>>> y = 3.42134234
>>> f'{y:0.4f}'
'3.4213'
```

## ※ 문자열 관련 함수

- 변수.count(’문자’) : 변수 안에 문자가 몇 개 있는지 카운트
- 변수.find(’문자’) : 변수 안에 문자가 처음 나온 위치를 인덱싱 숫자로 표현, 문자열 존재하지 않을 시 -1 반환
- 변수.index(’문자’) : 변수 안에 문자가 처음 나온 위치를 인덱싱 숫자로 표현, 문자열 존재하지 않을 시 오류 발생
- 문자열 삽입(join)

```python
>>>  ",".join('abcd')
'a,b,c,d'
```

→ abcd 문자열의 각각의 문자 사이에 , 기호 삽입

- 변수.upper() : 변수의 소문자를 대문자로 바꿔줌
- 변수.lower() : 변수의 대문자를 소문자로 바꿔줌
- 변수.lstrip() : 변수의 가장 왼쪽에 있는 한 칸 이상의 연속된 공백들을 모두 지움
- 변수.rstrip() : 변수의 가장 오른쪽에 있는 한 칸 이상의 연속된 공백들을 모두 지움
- 변수.strip() : 변수 양쪽에 있는 한 칸 이상의 연속된 공백들을 모두 지움
- 변수.replace(’바꿀 문자열’, ‘대체할 문자열’) : 변수의 바꿀 문자열을 대체할 문자열로 변경
- 변수.split() : 공백을 기준으로 문자열 나눔
- 변수.split(’기호’) : 기호를 기준으로 문자열 나눔
````
